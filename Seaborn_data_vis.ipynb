## Relplot

# Heatmap

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
pivote = flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
pivote
```

```python
sns.heatmap(pivote)
```

```python
# linewidth
sns.heatmap(pivote,linewidths=0.5)
```

```python
# annot fmt
sns.heatmap(pivote,annot=True,fmt='d')
```

# cluster map

```python
sns.clustermap(pivote)
```

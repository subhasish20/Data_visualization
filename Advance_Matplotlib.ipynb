```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
w = np.random.randint(1,4000,100)
x = np.random.randint(1,1000,100)
y = np.random.randint(1,2000,100)
z = np.random.randint(1,3000,100)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
plt.figure(figsize=(10,2))
plt.scatter(iris['sepal_length'],iris['petal_length'])
plt.show()
```

```python
iris.sample(5)
```

```python
iris['species'] = iris['species'].replace({'setosa':0,'versicolor':1,'virginica':2})
iris.sample(10)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(iris['sepal_length'],iris['petal_length'],c=iris['species'],cmap='viridis',s=100)
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y,c=z,s=200)
plt.scatter(x,y,c=w,s=200,cmap='cool')
plt.grid()
plt.colorbar()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,figsize=(10,5))
ax[0].scatter(x,y,c=z,s=200)
ax[1].scatter(y,z,c=w,cmap='plasma',s=200)
ax[0].grid()
ax[1].grid()
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
z = xx**2+yy**2
z.shape
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='rainbow')
fig.colorbar(p)
```

```python
z = np.sin(xx)+np.cos(yy)
fig = plt.figure(figsize=(10,6))
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

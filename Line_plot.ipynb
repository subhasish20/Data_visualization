# Line PLot

```python
from matplotlib import pyplot as plt
```

### Basic Line PLot

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [154,265,145,356,789,582,146,256]

#creating the plot
plt.plot(product_id, online_price)

plt.show()
```

## Labeling & changing colors
### use to lebel the sides 

```python
plt.plot(product_id, online_price,color="#1A37FF")

plt.xlabel("product_id")
plt.ylabel("online_price")
plt.title("product_id and online_price")

plt.show()
```

### chanding the labels colour

```python
plt.plot(product_id, online_price,color="#1A37FF")

plt.xlabel("product_id",color="blue")
plt.ylabel("online_price",color="blue")
plt.title("product_id and online_price",color="red")

plt.show()
```

## Markering the points and sizes changing
##### marker : add the marker
##### markersize : audjust the size of the 

```python
plt.plot(product_id, online_price,color="#1A37FF",marker = "+",markersize =10,)#marker and markersize

plt.xlabel("product_id")
plt.ylabel("online_price")
plt.title("product_id and online_price")

plt.show()
```

## Multiple values plot

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [154,265,145,356,789,582,146,256]
offline_price = [400,200,500,658,356,875,465,986]
chor_bazar = [123,231,321,152,146,256,361,245]

plt.plot(product_id,offline_price,color="green",marker="*",markersize=20)
plt.plot(product_id,online_price,color="blue",marker="*",markersize=20)
plt.plot(product_id,chor_bazar,color="red",marker="+",markersize=20)

plt.xlabel("product_id")
plt.ylabel("online_price")
plt.title("product_id and online_price")

plt.show()
```

## Lableing each single plot
### use to define the each plot in the graph

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [154,265,145,356,789,582,146,256]
offline_price = [400,200,500,658,356,875,465,986]
chor_bazar = [123,231,321,152,146,256,361,245]

plt.plot(product_id,offline_price,color="green",marker="*",markersize=20,label="offline_price")
plt.plot(product_id,online_price,color="blue",marker="*",markersize=20,label="online_price")
plt.plot(product_id,chor_bazar,color="red",marker="+",markersize=20,label="chor_bazar")

plt.xlabel("product_id")
plt.ylabel("online_price")
plt.title("product_id and online_price")

plt.legend() # use to show the labels

plt.show()
```

# Bar_Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Basic Bar_plot


```python
program = ["python","c","c++","java"]
popularity = [85,70,60,62]

plt.bar(program, popularity)

plt.show()
```

#### Labeling and changing the colors of bar and labels
#### Changing the colors of bar 

```python
program = ["python","c","c++","java"]
popularity = [85,70,60,62]

plt.bar(program, popularity, color="red")

plt.xlabel("program",color="red")
plt.ylabel("popularity",color="blue")
plt.title("programs and popularity charts",color="green") 

plt.show()
```

#### Changing the font size of the labels


```python
program = ["python","c","c++","java"]
popularity = [85,70,60,62]

plt.bar(program, popularity, color="red")

plt.xlabel("program",color="red",fontsize=20)
plt.ylabel("popularity",color="blue",fontsize=20)
plt.title("programs and popularity charts",color="green",fontsize=20) 

plt.show()
```

#### Changing the width size of the bars

```python
program = ["python","c","c++","java"]
popularity = [85,70,60,62]

plt.bar(program, popularity, color="blue",width=0.3) # width of the bars will be between 0 and 1

plt.xlabel("program",color="red")
plt.ylabel("popularity",color="blue")
plt.title("programs and popularity charts",color="green") 

plt.show()
```

#### Making the bars multiple colors

```python
program = ["python","c","c++","java"]
popularity = [85,70,60,62]

color_of_bars =["red","pink","blue","yellow"]

plt.bar(program, popularity, color=color_of_bars,width=0.3) # width of the bars will be between 0 and 1

plt.xlabel("program",color="red")
plt.ylabel("popularity",color="blue")
plt.title("programs and popularity charts",color="green") 

plt.show()
```

### Making the age colour in the bar chart & adjusting the edge size

```python
program = ["python","c","c++","java"]
popularity = [85,70,60,62]

color_of_bars =["red","pink","blue","yellow"]

plt.bar(program, popularity, color=color_of_bars,width=0.3,edgecolor="red",linewidth=5,linestyle=":") 
# the 'edgewidth' give the width and the 'linewidth' resized the edge and the linestyle changed the style of the plot

plt.xlabel("program",color="red")
plt.ylabel("popularity",color="blue")
plt.title("programs and popularity charts",color="green") 

plt.show()
```

### adjusting the color effects

```python
program = ["python","c","c++","java"]
popularity = [85,70,60,62]

color_of_bars =["red","pink","blue","yellow"]

plt.bar(program, popularity, color=color_of_bars,width=0.3,alpha=0.3) # alpha = 0.3 changed the color effect 

plt.xlabel("program",color="red")
plt.ylabel("popularity",color="blue")
plt.title("programs and popularity charts",color="green") 

plt.show()
```

### label parameter

```python
program = ["python","c","c++","java"]
popularity = [85,70,60,62]

plt.bar(program, popularity, color="blue",width=0.3,label="popularity charts") # width of the bars will be between 0 and 1

plt.xlabel("program",color="red")
plt.ylabel("popularity",color="blue")
plt.title("programs and popularity charts",color="green") 

plt.legend()
plt.show()
```

### Double bar plot

```python
program = ["python","c","c++","java"]
popularity = [85,70,60,62]
programmers = [70,50,45,30]

plt.bar(program, popularity, color="blue",width=0.3,label="popularity charts") # width of the bars will be between 0 and 1
plt.bar(program, programmers, color="green",width=0.3,label="popularity charts2")

plt.xlabel("program",color="red")
plt.ylabel("popularity",color="blue")
plt.title("programs and popularity charts",color="green") 

plt.legend()
plt.show()
```

```python
X = ['Group A','Group B','Group C','Group D']
Ygirls = [10,20,20,40]
Zboys = [20,30,25,30]
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Man',color='red')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Woman')
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Groups")
plt.title("Number of people in each group")
plt.legend()
plt.show()
```
